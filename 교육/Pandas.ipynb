### 행/열 접근
* loc : 이름으로 접근
* iloc : 순서로 접근

```python
# 0이라고 적힌 행에 접근
df.loc[0]

# 0,10,100,1000 이라고 적힌 행을 부름
df.loc[[0,10,100,1000]]

# 열 이름을 기입
df.loc[[0,10,100,1000], ['Channel','Term']]
```

```python
# 0번째 행에 접근
df.iloc[0]

# 0,10,100,1000 번째 행을 부름
df.iloc[[0,10,100,1000]]

#열 접근 때에도 열 순서를 기입
df.iloc[[0,10,100,1000], [4,6]] 
```

```python
# loc은 순서의 접근이 아니라 이름의 접근이라서 5를 포함해서 출력함
# 2와 5 사이에 2 0 7 14 5 라면 다 출력됨
df.loc[2:5,['Channel','Term']]
```

```python
# 하지만, iloc의 경우는 순서의 접근이라서 5를 포함하지 않고
# 정말로 2와 5 사이 (5제외) 값을 출력
df.iloc[2:5,[4,6]]
```

```python
# excel 파일을 불러올 때는 
# 파일을 열면, Sheet번호가 적혀있으면, sheet 번호를 명확하게 적어줘야함
df = pd.read_excel("실습 파일/성적 처리.xlsx", sheet_name='Sheet1')
df.head(2)
```

```python
df.iloc[:,2:].head(2)
```

```python
df['총합'] = df.iloc[:,2:].sum(axis=1)
df.head(3)
```

```python
df.sort_values("총합", ascending=False).head(10)
```

```python
df.sort_values(['반','총합'],ascending=False).head(10)
```

```python
df.sort_values(['반','총합'],ascending=[True,False]).head(10)
```

```python
# 순위 (기본값은 average), method 는 원래 생략되어 있음
# average 방법의 경우, 3등이 2명이면, 3.5룰 부여하고, 4등은 생략
df['순위'] = df['총합'].rank(method = "average",ascending=False)
df
```

```python
# max 방법의 경우, 3등이 2명이면, 모두 4등으로 처리하고, 3등은 생략
df['순위'] = df['총합'].rank(method = "max",ascending=False)
df
```

```python
# min 방법의 경우, 3등이 2명이면, 모두 3등으로 처리하고, 4등은 생략
df['순위'] = df['총합'].rank(method = "min",ascending=False)
df
```

```python
# first 방법의 경우, 3등이 2명이면, 먼저 발견한 사람이 3등이고 다음 발견된 사람이 4등으로 처리
# 선착순 만들때 주로 사용
df['순위'] = df['총합'].rank(method = "first",ascending=False)
df
```

```python
# dense 방법의 경우, min 방법처럼, 3등이 2명이면, 둘 다 3등으로 처리하고 
# 하지만, 다음 등수를 4등으로 처리 (min은 4등을 생략했음)
df['순위'] = df['총합'].rank(method = "dense",ascending=False)
df
```

```python
df1 = pd.read_csv("실습 파일/02_Data.csv")
df1.head(2)
```

```python
# 40대, 50대, 60대
def func1(x):
    if pd.isnull(x):
        return "확인 불가"
    
    if x >= 20 :
        return "성인"
    else:
        return "미성년자"
    
df1['성인판별']=df1['Age'].apply(func1)
df1
```
