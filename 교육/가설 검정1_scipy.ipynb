````python
def fac(x):
    if x==1 or x==0:
        return 1
    return x*fac(x-1)

print(fac(4))
````

````python
# lambda
voc_per_day = 5
time = 1
target = 0

(((voc_per_day)**target)*(np.exp(-voc_per_day)))/fac(target)

stats.poisson.pmf(target, voc_per_day*time)
````

````python
import pandas as pd
import numpy as np

df= pd.read_csv(" /01_Data.csv")
age = df['Age']
sns.histplot(age,kde=True)
plt.show()
````

````python
#  :   
#  :    .

# nan   NormaltestResult(statistic=nan, pvalue=nan)
stats.normaltest(age) 

stats.normaltest(age.dropna()) # H0  :    .
````

````python
# 1   (One Sample T-Test)

df['Datetime'] = pd.to_datetime(df['Datetime'])
df['Year'] = df.Datetime.dt.year

# 2023   
sample = df[df['Year']==2023]['Amount_Month']
sample
````

````python
#  : 2023     10 
#  : 2023     10 


stats.ttest_1samp(sample, 100000)# H0 ,     10 
````

````python
#        !!
stats.normaltest(sample) 
# H0 ,   .     
````

````python
#  10  
# 1   :        

import statsmodels.stats.descriptivestats as statsd

#  : 2023   10 
#  : 2023   10    

statsd.sign_test(sample, 100000) 
#   pvalue 
# H0 , 10    .
````
