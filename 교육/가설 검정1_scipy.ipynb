# 확증적 데이터 분석
확증적 데이터 분석(Confirmatory Data Analysis)은 데이터 분석에서 중요한 부분입니다. 이 접근법은 특정 가설을 설정하고, 그 가설을 데이터를 통해 검증하는 과정을 포함합니다. 확률분포의 이해는 이 과정에서 매우 중요합니다.



## 1. 확률분포의 역할
* 가설 설정: 연구자는 데이터에서 관찰된 패턴이나 현상에 대해 가설을 세웁니다.
* 통계적 가설검정: 이 가설을 검증하기 위해 통계적 가설검정을 사용합니다. 여기서는 ```귀무 가설(Null Hypothesis, H0)```과 ```대립 가설(Alternative Hypothesis, H1)```을 설정합니다.
* 확률분포 이해: 데이터의 확률분포를 이해하는 것은 중요합니다. 확률분포는 변수가 취할 수 있는 값과 그 값이 나타날 확률을 보여줍니다. 예를 들어, 정규분포는 종모양의 분포를 가지며, 많은 자연현상과 사회현상에서 관찰됩니다.
* 적절한 통계적 방법의 선택: 데이터의 분포에 따라 적절한 통계적 검정 방법을 선택해야 합니다. 예를 들어, 데이터가 정규분포를 따른다면 t-test나 ANOVA 같은 방법을 사용할 수 있습니다. 반면, 데이터가 정규분포를 따르지 않는다면 비모수적 방법을 사용해야 할 수 있습니다.

## 2. 연속 확률 분포
* 연속성: 계량형 확률 분포는 값이 특정 범위 안에서 어떤 값이든 취할 수 있는 연속적인 데이터를 다룹니다. 예를 들어, 사람의 ```키, 온도, 시간``` 등이 이에 해당합니다.
* 확률 밀도 함수(Probability Density Function, PDF): 연속 확률 분포는 확률 밀도 함수를 사용하여 표현됩니다. 이 함수는 특정 구간 내 데이터 값이 나타날 확률을 나타내며, 이 구간에 대한 확률 밀도 함수 아래의 면적이 그 구간에서 변수가 취할 수 있는 값의 확률을 나타냅니다.
* 누적 분포 함수(Cumulative Distribution Function, CDF): 누적 분포 함수는 확률 변수의 값이 특정 값보다 작거나 같을 확률을 나타내는 함수입니다. 이는 확률 밀도 함수를 이용하여 계산할 수 있습니다.

### 정규분포
* 정규분포는 연속 확률 분포의 한 종류로, 데이터가 평균(중심값) 근처에 밀집하여 대칭적인 종 모양의 분포를 이루는 것을 말합니다. 
* 이 분포는 좌우 대칭이며, 평균값을 중심으로 하여 좌우로 퍼져 나가는 모양을 가집니다. 평균값에 가까울수록 데이터의 밀도가 높아지고, 멀어질수록 밀도가 낮아집니다. 
* 정규분포는 자연과학, 사회과학, 경제학 등 다양한 분야에서 수집된 데이터의 분포를 근사하는 데 널리 사용됩니다. 이는 많은 현상들이 평균값을 중심으로 분포하는 경향을 보이기 때문입니다.

#### 중심극한정리(Central Limit Theorem)
* 독립적이고 동일하게 분포된 확률 변수들의 합이나 평균이, 그 변수의 개수가 충분히 크면 정규분포에 가까워진다는 정리입니다.
* 표본 평균이 모평균에 근사적으로 정규분포를 따른다는 것을 나타내며, 이로 인해 많은 통계적 방법론과 추정이 가능해집니다. 
* 표본의 크기가 큰 경우(일반적으로 30 이상)에는 모집단의 분포 형태와 관계없이 표본 평균의 분포가 정규분포를 따르는 것으로 간주됩니다.

#### 표준정규분포(Standard Normal Distribution)
* 평균이 0이고 표준편차가 1인 정규분포를 말합니다. 이는 모든 정규분포를 표준화하여 동일한 기준으로 비교할 수 있게 해줍니다.
* 어떤 정규분포든지 그 평균을 빼고 표준편차로 나누어 표준정규분포로 변환할 수 있습니다. 이 변환된 값을 Z-점수라고 합니다.
* Z = (X - μ) / σ, 여기서 X는 원래 변수, μ는 평균, σ는 표준편차입니다.

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 파이썬에서 많이 사용하는 통계적 계산 기능 라이브러리
import scipy.stats as stats

# 평균이 50이고, 표준편차가 10인 데이터 100개를 만든다
# np.random.normal(0,1, 갯수) 하면 표준정규분포를 만드는거
data = np.random.normal(50,10,100) 
data
````

### T 분포(T Distribution) 
정규 분포의 평균을 측정하기위해 고안된 분포로, 정규 분포와 유사하지만, 표본 크기에 따라 분포의 꼬리 부분이 더 두꺼운(tails are fatter) 모습을 가지고 있는 확률분포.
* 표준정규분포와 유사하게 0을 중심으로 좌우대칭
* 용도는 현재 가진 데이터를 표본으로 가정했을 때, 모집단의 평균을 추정하거나 검정할 때 사용

### 카이제곱분포(Chi²Distribution)
정규분포를 따르는 모집단에서 크기가 n인 표본을 무작위로 반복하여 추출하였을 때, 각 표본에 대해 구한 표본 분산들이 띄는 확률분포.
* 용도는 모집단의 분산 추정하거나, 빈도 기반의 분포 또는 형태 적합도 검정, 여러 집단 간의 독립성/동질성 검정 할 때 사용.

### F 분포(F Distribution)
분산이 같은 두 정규 모집단으로부터 크기 n1과 크기 n2의 확률표본을 반복하여 독립적으로 추출한 후, 구한 두 표본 분산의 비율들의 표본 확률 분포
* 용도는 F분포는 두 분포의 분산을 비교하는데 흔히 활용
* ANOVA에서는 그룹 내 변동과 그룹간 변동으로 여러 개의 평균값을 비교하는데 활용  
* 회귀분석에서 t분포는 개별 회귀계수의 유의성을 검정하는데 쓰이고, 회귀모형 자체의 유의성 검정하는데 사용

### 와이블 분포(Weibull Distribution)
지수분포를 보다 일반화시켜, 여러 다양한 확률분포 형태를 모두 나타낼 수 있도록 고안된 분포
* 스웨덴 물리학자 W. Weibull가 재료의 파괴강도를 설명하기 위해 1939년에 제안
* 용도는 신뢰성 분야에서 수명에 대한 추정을 하기 위해 활용

* 금융에서는 신용리스크 모델링, 수익률 평가, 금융파산 예측에 대해 해당 분포가 사용

## 3. 이산 확률 분포
계수형 확률 분포는 이산형 데이터를 모델링하는 데 사용되는 확률 분포입니다. 이러한 분포는 이산적인 결과를 가지는 확률 실험을 통해 나타나며, 각 결과에 대한 확률은 확률 질량 함수(Probability Mass Function, PMF)로 표현됩니다. 이산형 확률 분포의 대표적인 예로는 이항 분포(Binomial Distribution)와 포아송 분포(Poisson Distribution)가 있습니다.

### 이항분포(Binomial Distribution)
베르누이 시행(Bernoulli trial)이라고 불리는 실험을 n번 반복했을 때, 성공(또는 특정 결과)의 횟수가 따르는 분포입니다. 베르누이 시행은 두 가지 결과(성공/실패, 양품/불량 등)만을 가지는 실험입니다.
* 이 분포는 각 시행이 독립적이고, 각 시행에서 성공할 확률이 동일하다는 가정 하에 성립합니다.
* n번의 독립적인 베르누이 시행에서 성공 횟수 X가 따르는 확률 질량 함수는 P(X = k) = C(n, k) * p^k * (1-p)^(n-k)로 표현됩니다. 여기서 C(n, k)는 조합을 나타내며, p는 성공 확률입니다.

````python
# 동전을 3번 던졌을 때, 앞면이 한 번 나올 확률
# 3C2*(0.5)^1*(0.5)^2
````

### 포아송 분포 (Poisson Distribution)
* 단위 시간이나 단위 공간 내에서 발생하는 희귀 사건의 횟수를 나타내는 이산 확률 분포입니다. 
* 일정 시간 동안 발생하는 전화 수, 일정 면적의 철판에서 발견되는 결함 수, 특정 시간 동안 은행 창구를 방문하는 고객 수 등의 모델링에 적합합니다.
* 포아송 분포의 확률 질량 함수는 P(X = k) = (λ^k * e^(-λ)) / k!로 주어집니다. 여기서 λ는 단위 시간이나 단위 공간에서 기대되는 사건의 평균 발생 횟수입니다.

#### 포아송 분포 기대값 계산 문제
예제) 한 은행의 고객센터에 하루 평균 5번의 스마트 뱅크 사용 VOC 발생한다고 가정해보자. 특정 시간 동안 VOC 발생하지 않을 확률을 계산

````python
def fac(x):
    if x==1 or x==0:
        return 1
    return x*fac(x-1)

print(fac(4))
````

````python
# lambda
voc_per_day = 5
time = 1
target = 0

(((voc_per_day)**target)*(np.exp(-voc_per_day)))/fac(target)

stats.poisson.pmf(target, voc_per_day*time)
````

````python
person = 5
time = 1
target = 3

stats.poisson.pmf(target, person*time)
````

````python
target = 2
stats.poisson.pmf(0,person*time)+stats.poisson.pmf(1,person*time)+stats.poisson.pmf(2,person*time)
````

## 4. 추론 통계 (Inferencial Statistics)
추론 통계학(Inferential Statistics)은 표본 데이터를 이용하여 모집단에 대한 결론을 추론하는 통계학의 한 분야입니다. 기술 통계학(Descriptive Statistics)과 달리, 추론 통계학은 데이터의 요약 및 설명을 넘어 모집단에 대한 일반화된 결론을 내리는 데 초점을 맞춥니다.

### 통계적 추정(Statistical Estimation)
* 표본에서 얻은 통계량을 사용하여 모집단의 모수(Parameters)를 추정합니다. 예를 들어, 표본 평균을 사용하여 모집단의 평균을 추정할 수 있습니다.
* 추정에는 점추정(Point Estimation)과 구간추정(Interval Estimation)이 포함됩니다. 점추정은 모수의 단일 추정치를 제공하는 반면, 구간추정은 모수가 존재할 가능성이 있는 값의 범위를 제공합니다.

### 가설검정(Hypothesis Testing)
* 가설검정은 모집단에 대한 특정 가설을 세운 후, 표본 데이터를 이용하여 그 가설의 타당성을 검증하는 과정입니다.
* 일반적으로 귀무 가설(Null Hypothesis, H0)과 대립 가설(Alternative Hypothesis, H1)을 설정하고, 표본 데이터를 분석하여 귀무 가설을 기각하거나 채택합니다.

### 이론적 매커니즘
* 모수(Parameter): 모집단의 특성을 기술하는 값입니다. 예를 들어, 모평균, 모분산 등이 있습니다.
* 통계량(Statistic): 표본 데이터로부터 계산되는 값으로, 모집단의 모수를 추정하기 위해 사용됩니다. 예를 들어, 표본 평균, 표본 분산 등이 있습니다.

### 신뢰구간(Confidence Interval)과 신뢰수준(Confidence Level)
* 신뢰구간: 모수가 존재할 것으로 추정되는 값의 범위입니다. 신뢰구간은 일반적으로 점추정치 주변에 설정되며, 오차 범위를 고려하여 계산됩니다.
* 신뢰수준: 신뢰구간이 모수를 포함할 확률을 나타냅니다. 예를 들어, 95% 신뢰수준에서의 신뢰구간은 같은 방법으로 추출된 표본에서 구한 신뢰구간 중 약 95%가 실제 모수를 포함할 것으로 예상됩니다.
* 유의수준(Significance Level): 귀무 가설을 잘못 기각할 최대 허용 오류율을 의미합니다. 일반적으로 0.05(5%)를 사용합니다.


## 5. 통계적 가설 검정
통계적 가설 검정은 데이터를 통해 특정 가설이 참인지 거짓인지를 판단하는 과정입니다. 이 과정은 두 가지 주요 가설, 즉 귀무가설(Null Hypothesis)과 대립가설(Alternative Hypothesis)을 설정하고 이들 사이에서 결론을 도출하는 방식으로 진행됩니다.

#### 귀무가설 (Null Hypothesis, H0)
* 귀무가설은 현재까지 주장되어 온 것이거나 변화나 차이가 없다고 가정하는 기본적인 가설입니다.
* 일반적으로 "평균이 같다", "연관성이 없다", "독립적이다"와 같은 상태를 가정합니다.
* 가설 검정의 목적은 대부분 귀무가설을 기각하는 것입니다.

#### 대립가설 (Alternative Hypothesis, H1)
* 대립가설은 새로운 주장을 나타내며, 귀무가설과는 반대되는 가설입니다.
* "평균이 다르다", "연관성이 있다", "상관관계가 있다"와 같이 귀무가설에 반대되는 상태를 가정합니다.
* 귀무가설이 기각될 경우, 대립가설이 채택됩니다.

#### 가설 검정의 기본 가정
* 독립적인 데이터 수집: 수집된 데이터는 서로 영향을 미치지 않고 독립적으로 수집되었다고 가정합니다.
* 모집단에서의 표본 추출: 표본은 모집단에서 무작위로 추출된 것으로 가정합니다. 이는 표본이 모집단을 대표한다고 보는 것입니다.
* 두 개의 표본 비교 시 동일 모집단 추출 가정: 두 표본을 비교하는 경우, 이들이 동일한 모집단에서 추출되었다고 가정합니다.

#### 가설 검정 과정
* 귀무가설과 대립가설 설정: 연구 질문에 따라 귀무가설과 대립가설을 설정합니다.
* 적절한 통계적 검정 선택: 연구 질문과 데이터의 종류(연속형, 범주형 등)에 따라 적절한 통계적 검정 방법을 선택합니다.
* 검정 통계량 계산 및 결론 도출: 선택된 검정 방법을 사용하여 검정 통계량을 계산하고, 이를 기반으로 귀무가설을 기각하거나 채택합니다.

#### 신뢰수준과 유의수준
* 신뢰수준(Confidence Level): 가설 검정 시, 귀무가설을 기각하는 결정이 올바를 확률을 나타냅니다. 일반적으로 95% 또는 99%가 사용됩니다.
* 유의수준(Significance Level): 귀무가설을 잘못 기각할 최대 허용 오차율을 나타내며, 보통 0.05(5%) 또는 0.01(1%)를 사용합니다.

#### α-Risk와 β-Risk
* 실제로 가설 검정을 수행하게 되면, 가설검정의 결과가 실제 벌어지는 상황과 다른 경우가 있을 수 있습니다.
* 이유는 가설검정이 현재 수집된 데이터를 표본으로 가정하고, 모집단을 추정해 가설이 참인지 거짓인지 결론을 내리는 것이기 때문입니다.
    * α-Risk : 가설검정의 결과는 대립가설이 채택 되었으나, 실제로는 귀무가설이 참인 경우
    * β-Risk : 가설 검정의 결과는 귀무가설이 채택 되었으나, 실제로는 대립가설이 참인 경우
    


* 검정력(Power of the test) : 귀무가설이 틀릴 경우 귀무가설을 기각할 확률 즉, 틀린 가설을 틀렸다고 올바르게 판정할 확률을 의미
* P value (확률 값) : 귀무가설이 참일 확률
* 유의 수준 (0.05 / 5%) : 가설검정의 판단 기준

* 귀무가설 Ho가 참일 때 귀무가설 Ho를 기각할 오류를 범할 확률의 최대 허용한계 (제 1종 오류의 최대치)  
    * P.value > 0.05 (5%) : 귀무가설 참 (귀무가설기각실패)  
    * P.value < 0.05 (5%) : 대립가설 참 (귀무가설기각)  

Ex) 10대와 20대의 주당방문횟수의 평균의 차이가 있는가?  
* 귀무 가설 : 10대와 20대 집단 간의 평균의 유의미한 차이가 없다.
* 대립 가설 : 10대와 20대 집단 간의 평균의 유의미한 차이가 있다!
* P.value가 만약 3%로 계산 된다면, 통계의 일반적 유의수준 5%보다 작으므로, 대립가설 참 (귀무가설기각)
* 사실적 결론 : 10대와 20대 집단 간의 평균의 유의미한 차이가 있다!
