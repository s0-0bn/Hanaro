## Pyplot
matplotlib.pyplot 모듈은 MATLAB과 비슷하게 명령어 스타일로 동작하는 함수의 모음입니다.
matplotlib.pyplot 모듈의 각각의 함수를 사용해서 간편하게 그래프를 만들고 변화를 줄 수 있습니다.
예를 들어, 그래프 영역을 만들고, 몇 개의 선을 표현하고, 레이블로 꾸미는 등의 일을 할 수 있습니다.

````python
import matplotlib.pyplot as plt
import numpy as np
import koreanize_matplotlib
````

### 1. 기본 그래프 그리기
* pyplot.plot() 함수에 하나의 숫자 리스트를 입력함으로써 아래와 같은 그래프가 그려집니다.
* plot() 함수는 리스트의 값들이 y 값들이라고 가정하고, x 값 [0, 1, 2, 3]을 자동으로 만들어냅니다.
* matplotlib.pyplot 모듈의 show() 함수는 그래프를 화면에 나타나도록 합니다.

````python
x = [1,2,3,4]
plt.plot(x)
plt.show()
````

* plot() 함수는 다양한 기능을 포함하고 있어서, 임의의 개수의 인자를 받을 수 있습니다. 예를 들어, 아래와 같이 입력하면, x-y 값을 그래프로 나타낼 수 있습니다.

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y)
plt.show()
````

### 2. 스타일 지정하기
matplotlib.pyplot의 plot() 함수는 다양한 옵션을 제공하여 그래프의 스타일과 형태를 사용자가 원하는 대로 설정할 수 있게 해줍니다.

* color 또는 c: 선의 색상을 설정합니다. 색상 이름이나 RGB 코드 등을 사용할 수 있습니다.  
예: color='blue', color='#FFDD44', color=(1.0,0.2,0.3)

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y, color='#CC33CC')
plt.show()
````

* linestyle 또는 ls: 선의 스타일을 설정합니다. 예를 들어, -, --, -., : 등이 있습니다.  
예: linestyle='-' (실선), linestyle='--' (파선)

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y, color='red', linestyle="-.")
plt.show()
````

* linewidth 또는 lw: 선의 너비를 설정합니다. 일반적으로 실수 값을 사용합니다.  
예: linewidth=2.5

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y, color='red', 
         linestyle="-.", 
         linewidth='2')
plt.show()
````

* marker: 데이터 포인트마다 표시할 마커의 종류를 설정합니다. 예를 들어, o, ^, s, *,D 등이 있습니다.  
예: marker='o' (원형 마커), marker='^' (삼각형 마커)

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y, color='red', 
         linestyle="-.", 
         linewidth='2',
        marker='o')
plt.show()
````

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y,'*')
plt.show()
````

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y,'r*:')
plt.show()
````

* markersize 또는 ms: 마커의 크기를 설정합니다.  
예: markersize=10

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y, color='red', 
         linestyle="-.", 
         linewidth='2',
        marker='o', markersize=10)
plt.show()
````

* markeredgecolor 또는 mec: 마커의 테두리 색상을 설정합니다.  
예: markeredgecolor='black'

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y, 
         color='red', linestyle="-.", linewidth='1',
        marker='o', markersize=10, markeredgecolor='black')
plt.show()
````

* markeredgewidth 또는 mew: 마커의 테두리 너비를 설정합니다.  
예: markeredgewidth=1.5

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y, 
         color='yellow', linestyle="-.", linewidth='1',
        marker='o', markersize=10, markeredgecolor='black',
        markeredgewidth=2)
plt.show()
````

* markerfacecolor 또는 mfc: 마커 내부의 색상을 설정합니다.  
예: markerfacecolor='red'

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y, 
         color='yellow', linestyle="-.", linewidth='1',
        marker='o', markersize=10, markeredgecolor='black',
        markeredgewidth=3, markerfacecolor='red')
plt.show()
````

* alpha: 선의 투명도를 설정합니다. 0(완전 투명)에서 1(완전 불투명) 사이의 값을 가집니다.  
예: alpha=0.5

````python
y = [1,4,8,16]
x = [1,2,3,4]
plt.plot(x,y, 
         color='yellow', linestyle="-.", linewidth='1',
        marker='o', markersize=10, markeredgecolor='black',
        markeredgewidth=3, markerfacecolor='red', alpha = 0.5)
plt.show()
````

### 3. 축 범위 설정
* axis([xmin, xmax, ymin, ymax]) 형태로 사용하여 x축과 y축의 최소값과 최대값을 설정할 수 있습니다.  
(예: plt.axis([0, 10, 0, 20])은 x축을 0부터 10까지, y축을 0부터 20까지 표시합니다.)

* xlim() - X축이 표시되는 범위를 지정하거나 반환합니다.
* ylim() - Y축이 표시되는 범위를 지정하거나 반환합니다.

````python
# plt.axis([0,10,0,20])
plt.plot(x,y)

a = plt.xlim(0,10)
b = plt.ylim(0,20)

plt.xlabel('X축')
plt.ylabel('Y축')

plt.show()
````

````python
a,b
````

* 'on' / 'off': 축의 눈금과 레이블을 표시하거나 숨길 수 있습니다. 'on'은 축을 표시하고, 'off'는 축을 숨깁니다.  
예: plt.axis('off')는 모든 축 눈금과 레이블을 숨깁니다.

````python
# plt.axis([0,10,0,20])
plt.plot(x,y)

a = plt.xlim(0,10)
b = plt.ylim(0,20)

plt.xlabel('X축')
plt.ylabel('Y축')

plt.axis('off')

plt.show()
````

* 'equal': x축과 y축의 스케일을 동일하게 설정하여 그래프가 왜곡되지 않도록 합니다.  
예: plt.axis('equal')

````python
# plt.axis([0,10,0,20])
plt.plot(x,y)

a = plt.xlim(0,10)
b = plt.ylim(0,20)

plt.xlabel('X축')
plt.ylabel('Y축')

plt.axis('equal')

plt.show()
````

* 'square': 축 범위를 제어하지 않고, 그래프 영역을 정사각형으로 만듭니다.

````python
# plt.axis([0,10,0,20])
plt.plot(x,y)

a = plt.xlim(0,10)
b = plt.ylim(0,20)

plt.xlabel('X축')
plt.ylabel('Y축')

plt.axis('square')

plt.show()
````

* 'tight': 모든 데이터가 그래프에 꽉 차게 보이도록 축 범위를 조정합니다.  
예: plt.axis('tight')

````python
# plt.axis([0,10,0,20])
plt.plot(x,y)

a = plt.xlim(0,10)
b = plt.ylim(0,20)

plt.xlabel('X축')
plt.ylabel('Y축')

plt.axis('tight')

plt.show()
````

* 'scaled': 'equal'과 비슷하지만, 축의 길이 비율이 데이터의 범위 비율과 동일하게 조정됩니다.

````python
# plt.axis([0,10,0,20])
plt.plot(x,y)

a = plt.xlim(0,10)
b = plt.ylim(0,20)

plt.xlabel('X축')
plt.ylabel('Y축')

plt.axis('scaled')

plt.show()
````

### 4. 여러 개의 그래프 그리기
* Matplotlib에서는 일반적으로 NumPy 어레이를 이용하게 되는데, 사실 NumPy 어레이를 사용하지 않더라도 모든 시퀀스는 내부적으로 NumPy 어레이로 변환됩니다.

#### 색상 (Color)
```c``` (청록색, cyan), ```m``` (자홍색, magenta), ```y``` (노란색, yellow), ```k``` (검정색, black), ```w``` (흰색, white) 등이 있습니다.

#### 마커 (Marker)
```o``` (원형), ```*``` (별표), ```.``` (점), ```x``` (엑스자), ```D``` (다이아몬드), ```h``` (육각형), ```^``` (삼각형) 등이 있습니다.

#### 선 스타일 (Linestyle)
```-``` (Solid), ```--``` (Dashed), ```:``` (Dotted), ```-.``` (Dash-dot)과 같이 네가지 종류를 선택할 수 있습니다.

````python
x = np.arange(2,5,0.5)

plt.plot(x,x,'-.s')
plt.plot(x,x**2,':g*')
plt.plot(x,x**3,'--r^')

plt.show()
````

### 5. 축 레이블 설정하기
matplotlib.pyplot 모듈의 xlabel(), ylabel() 함수를 사용하면 그래프의 x, y 축에 대한 레이블을 표시할 수 있습니다.

````python
x = np.arange(2,5,0.5)

plt.plot(x,x,'-.s')
plt.plot(x,x**2,':g*')
plt.plot(x,x**3,'--r^')

plt.xlabel("X")
plt.ylabel("Y")

plt.show()
````

* 여백 지정하기
xlabel(), ylabel() 함수의 labelpad 파라미터는 축 레이블의 여백 (Padding)을 지정합니다.

````python
x = np.arange(2,5,0.5)

plt.plot(x,x,'-.s')
plt.plot(x,x**2,':g*')
plt.plot(x,x**3,'--r^')

plt.xlabel("X", labelpad=5)
plt.ylabel("Y", labelpad=25)

plt.show()
````
