### merge
데이터 합병

```python
# 데이터 병합
df_member = pd.read_csv("실습 파일/03_store_member.csv")
df_product = pd.read_csv("실습 파일/03_store_product.csv")
```

```python
df_member.head(2)
```

```python
df_product.head(2)
```

```python
p1 = df_product.pivot_table(index="회원번호", values=['구매금액', '구매수량'], aggfunc = 'sum').reset_index()
p1.info()
```

```python
p1['회원번호'] = p1['회원번호'].astype('int')
p1 = p1.rename(columns={'구매금액':'총구매금액','구매수량':'총구매수량'})
p1
```

```python
# 두 개의 데이터프레임을 합치기
# pd.merge(df_member, p1)을 해도 똑같이 나와유
# df_member.merge(p1, on = "회원번호") 컬럼명이 동일한 경우는 그냥 on 도 사용가능
# 그래도~ 다른 경우가 많으니깐 다르게 적장
m1 = df_member.merge(p1, left_on = "회원번호", right_on = "회원번호")
m1
```

```python
# 기준열이 같을 땐, on = 기준 이름
# how : outer(양쪽 데이터 그냥 다 합쳐라), 
#      inner(기본값), 
#      left(왼쪽 데이터 프레임 기준)
#      right(오른쪽 데이터 프레임 기준)
m1 = df_member.merge(p1, left_on='회원번호', right_on = '회원번호', how = 'outer')
m1
```

```python
df_product
```

```python
# 한 명의 회원이 몇 번 방문했는가 계산하기
df_product['Label']=1
p3 = df_product.pivot_table(index=['회원번호', '공급일자'], values="Label", aggfunc="min").reset_index()
p3 = p3.pivot_table(index="회원번호", values="Label", aggfunc= "sum").reset_index()
p3 = p3.rename(columns = {"Label":"총방문횟수"})
p3
```

```python
m1.merge(p3, on = "회원번호", how='outer')
```

```python
# 한 명의 회원이 몇 번 방문했는가 계산하기
sample = pd.crosstab(index=df_product["회원번호"], columns=df_product["공급일자"]).reset_index()
pd.melt(sample, id_vars="회원번호")
```

```python
pd.crosstab(index=df_product["회원번호"], columns=df_product["공급일자"], margins=True)
```

```python
pd.melt(sample, id_vars="회원번호")
```

### 그래프 그리기

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# dataset별로 x,y의 평균과 표준편차 구하기
anscombe = sns.load_dataset("anscombe")
anscombe.pivot_table(index='dataset' ,aggfunc=['mean','std']).reset_index()
```

```python
data1 = anscombe[anscombe['dataset']=='I']
data2 = anscombe[anscombe['dataset']=='II']
data3 = anscombe[anscombe['dataset']=='III']
data4 = anscombe[anscombe['dataset']=='IV']
```

```python
print(data1['x'].corr(data1['y']))
print(data2['x'].corr(data2['y']))
print(data3['x'].corr(data3['y']))
print(data4['x'].corr(data4['y']))
```

```python
# dataset별로 평균과 표준편차가 비슷하게 보이더라도
# 산점도를 그렸을 때, 다 다른 경향을 보일 수 있음!
# 주의주의

fig = plt.figure() #흰색 도화지

ax1 = fig.add_subplot(2,2,1) #2행 2열의 1번째
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.scatter(data1['x'],data1['y'])
ax2.scatter(data2['x'],data2['y'])
ax3.scatter(data3['x'],data3['y'])
ax4.scatter(data4['x'],data4['y'])

ax1.set_title("Data1")
ax2.set_title("Data2")
ax3.set_title("Data3")
ax4.set_title("Data4")

fig.suptitle("Anscombe Data")
fig.tight_layout()

plt.savefig("실습 파일/Anscombe.png")
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
#히스토그램(막대 그래프 : 일변량 그래프 - 연속변수)
sns.set_style("whitegrid")
sns.histplot(data=tips, x='total_bill', bins=20, kde=True) #밀도 함수 자동으로 그려줌
# sns.histplot(data=tips, x='tip', bins=10) # 동시에 작성 가능
```

```python
# 산점도(이변량 그래프 - 수치형 변수)
sns.lmplot(data=tips, x='total_bill', y='tip', fit_reg=True) #회귀선 그리기 fit_reg
```

```python
# 빈도 막대 그래프 (일변량 그래프 - 범주형 변수)
sns.countplot(data=tips, x='day')
```

```python
# 평균 막대 그래프 (이변량 그래프 : x축이 범주형, y축 수축형)
# 물론 역도 가능
sns.barplot(data=tips, x='sex', y='tip')
```

```python
# 박스 플롯 : 이변량 그래프(x축은 범주형, y축은 수치형)
sns.boxplot(data=tips, x='sex', y='tip')
```

```python
# 바이올린 플롯 : 이상치가 몇갠지 몰라서 잘 사용하진 않음
# 스왐플롯
sns.violinplot(data=tips, x='sex', y='tip')
sns.swarmplot(data=tips, x='sex', y='tip', color = 'white')
```

```python
# 히스토그램 + 산점도
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex', color='blue')
```

```python
sns.kdeplot(data=tips, x='total_bill', y = 'tip', fill=True, color = 'lightgrey')
```

```python
#밀도 함수 그래프 : 일변량, 이변량
sns.lmplot(data=tips, x='total_bill',y='tip', fit_reg=False)
sns.kdeplot(data=tips, x='total_bill',y='tip', colors ='orange')
```

```python
sns.lmplot(data=tips, x='total_bill',y='tip', fit_reg=False,
          hue='sex') # hue는 주로 범주형으로 3변량으로 만들어줌
```

```python
sns.lmplot(data=tips, x='total_bill',y='tip', fit_reg=False,
          hue='sex', col='time') # col는 주로 범주형으로 4변량으로 만들어줌
```

```python
sns.lmplot(data=tips, x='total_bill',y='tip', fit_reg=False,
          hue='sex', col='time', row = 'smoker') # col는 주로 범주형으로 5변량으로 만들어줌
```
