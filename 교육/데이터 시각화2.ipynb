## Pyplot
matplotlib.pyplot 모듈은 MATLAB과 비슷하게 명령어 스타일로 동작하는 함수의 모음입니다.
matplotlib.pyplot 모듈의 각각의 함수를 사용해서 간편하게 그래프를 만들고 변화를 줄 수 있습니다.
예를 들어, 그래프 영역을 만들고, 몇 개의 선을 표현하고, 레이블로 꾸미는 등의 일을 할 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np
import koreanize_matplotlib
```

### 1. 폰트 설정하기
xlabel(), ylabel() 함수의 fontdict 파라미터를 사용하면 축 레이블의 폰트 스타일을 설정할 수 있습니다.

* family: 글꼴 가족을 설정합니다. 예를 들어, 'serif', 'sans-serif', 'cursive', 'fantasy', 'monospace' 등이 있습니다.
* color: 글꼴 색상을 설정합니다. 색상은 이름이나 RGB 코드로 지정할 수 있습니다.
* weight: 글꼴의 두께를 설정합니다. 예를 들어, 'normal', 'bold', 'heavy', 'light', 'ultrabold', 'ultralight' 등이 있습니다.
* size: 글꼴 크기를 설정합니다. 정수나 'smaller', 'x-large'와 같은 상대적 크기를 지정할 수 있습니다.
* style: 글꼴 스타일을 설정합니다. 'normal', 'italic', 'oblique' 등이 있습니다.
* backgroundcolor: 텍스트의 배경 색상을 설정합니다.
* alpha: 텍스트의 투명도를 설정합니다. 0(완전 투명)부터 1(완전 불투명) 사이의 값을 지정할 수 있습니다.

```python
x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y)

plt.xlabel("X-Axis", fontdict={'family':'serif',"color":"blue",
                              "weight":'bold', 'size':14})

plt.ylabel("y-Axis", fontdict={'family':'cursive',"color":"deeppink",
                              "weight":'normal', 'size':'x-large'})

plt.show()
```

### 2. 위치 지정하기
* xlabel() 함수의 loc 파라미터는 X축 레이블의 위치를 지정합니다. ({‘left’, ‘center’, ‘right’})
* ylabel() 함수의 loc 파라미터는 Y축 레이블의 위치를 지정합니다. ({‘bottom’, ‘center’, ‘top’})

```python
x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y)

plt.xlabel("X-Axis", loc = "right")
plt.ylabel("y-Axis", loc = "top")

plt.show()
```

### 3. 범례 표시하기
그래프 영역에 범례를 나타내기 위해서는 우선 plot() 함수에 label 문자열을 지정하고, matplotlib.pyplot 모듈의 legend() 함수를 호출합니다.

```python
x = [1,2,3,4]
y1 = [1,4,9,16]
y2 = [1,3,5,7]

plt.plot(x,y1,label= 'Price')
plt.plot(x,y2,label= "Price2")

plt.xlabel("X-Axis", loc = "right")
plt.ylabel("y-Axis", loc = "top")

plt.legend()

plt.show()
```

#### 범례 위치 조정하기

legend() 함수의 loc 파라미터를 이용해서 범례가 표시될 위치를 설정할 수 있습니다. 
* loc 파라미터를 숫자 쌍 튜플로 지정하면, 해당하는 위치에 범례가 표시됩니다. 
* loc=(0.0, 0.0)은 데이터 영역의 왼쪽 아래, loc=(1.0, 1.0)은 데이터 영역의 오른쪽 위 위치입니다.

```python
x = [1,2,3,4]
y1 = [1,4,9,16]
y2 = [1,3,5,7]

plt.plot(x,y1,label= 'Price')
plt.plot(x,y2,label= "Price2")

plt.xlabel("X-Axis", loc = "right")
plt.ylabel("y-Axis", loc = "top")

# 왼쪽 하단을 중심으로 위치함.
# plt.legend(loc = (0.5,0.5))
plt.legend(loc = (1,1)) 

plt.show()
```

loc 파라미터는 예제에서와 같이 문자열로 지정할 수도 있고, 숫자 코드를 사용할 수도 있습니다. loc=’lower right’와 같이 지정하면 아래와 같이 오른쪽 아래에 범례가 표시됩니다.

* 'best': 데이터를 가리지 않는 최적의 위치를 자동으로 선택합니다.
* 'upper right': 그래프의 오른쪽 상단에 위치합니다.
* 'upper left': 그래프의 왼쪽 상단에 위치합니다.
* 'lower left': 그래프의 왼쪽 하단에 위치합니다.
* 'lower right': 그래프의 오른쪽 하단에 위치합니다.
* 'right': 그래프의 오른쪽에 위치합니다.
* 'center left': 그래프의 왼쪽 중앙에 위치합니다.
* 'center right': 그래프의 오른쪽 중앙에 위치합니다.
* 'lower center': 그래프의 아래쪽 중앙에 위치합니다.
* 'upper center': 그래프의 위쪽 중앙에 위치합니다.
* 'center': 그래프의 중앙에 위치합니다.

```python
x = [1,2,3,4]
y1 = [1,4,9,16]
y2 = [1,3,5,7]

plt.plot(x,y1,label= 'Price')
plt.plot(x,y2,label= "Price2")

plt.xlabel("X-Axis", loc = "right")
plt.ylabel("y-Axis", loc = "top")

plt.legend(loc = 'lower right')

plt.show()
```

#### 범례 열 개수 조정하기
legend() 함수의 ncol 파라미터는 범례에 표시될 텍스트의 열의 개수를 지정합니다.

```python
x = [1,2,3,4]
y1 = [1,4,9,16]
y2 = [1,3,5,7]

plt.plot(x,y1,label= 'Price')
plt.plot(x,y2,label= "Price2")

plt.xlabel("X-Axis", loc = "right")
plt.ylabel("y-Axis", loc = "top")

plt.legend(ncol = 2)

plt.show()
```

#### 폰트 크기 지정하기
legend() 함수의 fontsize 파라미터는 범례에 표시될 폰트의 크기를 지정합니다.

```python
x = [1,2,3,4]
y1 = [1,4,9,16]
y2 = [1,3,5,7]

plt.plot(x,y1,label= 'Price')
plt.plot(x,y2,label= "Price2")

plt.xlabel("X-Axis", loc = "right")
plt.ylabel("y-Axis", loc = "top")

plt.legend(ncol = 2, fontsize=14)

plt.show()
```

#### 범례 테두리 꾸미기
frameon 파라미터는 범례 텍스트 상자의 테두리를 표시할지 여부를 지정합니다.
frameon=False로 지정하면 테두리가 표시되지 않습니다.
shadow 파라미터를 사용해서 텍스트 상자에 그림자를 표시할 수 있습니다.

( 이 외에도 legend() 함수에는 facecolor, edgecolor, borderpad, labelspacing과 같은 다양한 파라미터가 있습니다. 더 자세한 내용은 공식 홈페이지의 설명 ( https://matplotlib.org/stable/ )을 참고하세요. )

```python
x = [1,2,3,4]
y1 = [1,4,9,16]
y2 = [1,3,5,7]

plt.plot(x,y1,label= 'Price')
plt.plot(x,y2,label= "Price2")

plt.xlabel("X-Axis", loc = "right")
plt.ylabel("y-Axis", loc = "top")

plt.legend(ncol = 2, frameon=False)

plt.show()
```

```python
x = [1,2,3,4]
y1 = [1,4,9,16]
y2 = [1,3,5,7]

plt.plot(x,y1,label= 'Price')
plt.plot(x,y2,label= "Price2")

plt.xlabel("X-Axis", loc = "right")
plt.ylabel("y-Axis", loc = "top")

plt.legend(ncol = 2, shadow=True)

plt.show()
```

### 4. 그래프 영역 채우기
* fill_between() - 두 수평 방향의 곡선 사이를 채웁니다.

```python
x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y)
plt.fill_between(x[1:3], y[1:3], alpha=0.5)
plt.show()
```

* fill_betweenx() - 두 수직 방향의 곡선 사이를 채웁니다.

```python
x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y)
plt.fill_betweenx(y[:2],x[:2], alpha=0.5)
plt.show()
```

```python
x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,7,8]

plt.plot(x,y1,color = 'skyblue')
plt.plot(x,y2, color = 'gray')

plt.fill_between(x[1:3],y1[1:3],y2[1:3], 
                 color = 'pink',alpha=0.5)

plt.show()
```

* fill() - 다각형 영역을 채웁니다.

```python
x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,7,8]

plt.plot(x,y1)
plt.plot(x,y2)

plt.fill([2,2,3,3],[2,4,4,2], color = "lightgray" ,alpha=0.5)

plt.show()
```

### 5. 그리드 설정하기
데이터의 위치를 더 명확하게 나타내기 위해 그래프에 그리드 (Grid, 격자)를 표시할 수 있습니다.

```python
x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,7,8]

plt.plot(x,y1)
plt.plot(x,y2)

plt.fill([2,2,3,3],[2,4,4,2], color = "lightgray" ,alpha=0.5)

plt.grid()
plt.show()
```

#### 축 지정하기
axis=y로 설정하면 가로 방향의 그리드만 표시됩니다.
* {‘both’, ‘x’, ‘y’} 중 선택할 수 있고 디폴트는 ‘both’입니다.

```python
x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,7,8]

plt.plot(x,y1)
plt.plot(x,y2)

plt.fill([2,2,3,3],[2,4,4,2], color = "skyblue" ,alpha=0.5)

plt.grid(axis='y', linestyle=':')
plt.show()
```

#### 스타일 지정하기
color, alpha, linestyle 파마리터를 사용해서 그리드 선의 스타일을 설정했습니다.

```python
x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,7,8]

plt.plot(x,y1)
plt.plot(x,y2)

plt.fill([2,2,3,3],[2,4,4,2], color = "pink")

plt.grid(axis='x', alpha = 0.5, color='blue', linestyle='-.')
plt.show()
```
