## 1. 주어진 지표만 활용하여 저평가 주식 선별

### a. 수치형 변수 0 이상 만들기

```python
indicies = []

temp = stock.drop(['종목명','등락률','소속'],axis=1)

for idx,values in temp.iterrows():
    for j in values:
        if j < 0:
            indicies.append(idx)
            break            
```

```python
stock = stock.drop(indicies,axis=0)
```

### b. PER & PBR
- PER
    - 기업의 주가를 기준으로 이익을 얼마나 높게 평가하고 있는지를 나타내는 지표. 
    - 일반적으로 낮은 PER을 가진 기업은 저평가되어 있다고 볼 수 있음
    - **일반적으로 10을 기준으로 보지만, 산업에 따라 15를 볼 수도 있어서 기준을 15로 선정**

- PBR
    - 기업의 주가를 순자산가치로 나눈 비율. 
    - PBR이 1보다 낮을수록 주식이 저평가되었다고 볼 수 있음

```python
stock2 = stock[['종목명','현재가','거래량','상장주식수','PER','PBR','ROE','ROA','매출액증가율','부채총계','자산총계','유보율','소속']]
sample = stock2[(stock2['PBR']<1)&(stock2['PER']<15)]
sample.head(3)
```

### c. ROE & ROA
- ROE(자기자본이익률)
    - 기업의 순이익을 자본에 대한 비율로 나타내는 지표 
    - 높은 ROE는 기업이 효율적으로 자본을 운용한다는 것을 나타냄

- ROA(총자산수익률)
    - 기업의 순이익을 총 자산에 대한 비율로 나타내는 지표
    - 높은 ROA는 기업이 자산을 효율적으로 활용한다는 것을 나타냄
    
**ROA와 ROE는 높을수록 좋지만, 이 두 지표에 큰 차이를 보이는 것은 좋지 않음. 때문에, 이 둘의 차이가 7 미만인 기업을 선별**

```python
sample2 = sample.sort_values(by = ['ROE','ROA','매출액증가율'], ascending=[False,False,False])
sample3 = sample2[abs(sample2['ROE']-sample2['ROA'])<7]
sample3.head(4)
```

### d. 부채비율
- 부채비율
    - 부채총계/자산총계
    - 부채비율이 50% 미만인 기업들을 선별
    - 부채비율이 50% 이상인 경우, 자본보다 부채가 더 크기에 50으로 선정

```python
sample3['부채비율'] = sample3['부채총계']/sample3['자산총계']
sample4 = sample3[['종목명', '현재가', '거래량', '상장주식수', 'PER', 'PBR', 'ROE', 'ROA', '매출액증가율','부채비율','부채총계','자산총계','유보율','소속']]
sample5 = sample4[sample4['부채비율']<0.5]
sample5
```

## 2. 순자산가치법 적용하여 저평가 주식 선별
- 순자산가치법(Net Asset Value, NAV) : 순자산가치법은 기업의 순자산(자산 - 부채)을 기준으로 주식의 가치를 평가
    - 주식의 현재 가치 = 기업의 순자산 / 발행된 주식 수

```python
# 단위 동일하게 맞추기
sample5['자산총계'] = sample5['자산총계'] * 100000000
sample5['부채총계'] = sample5['부채총계'] * 100000000
sample5['상장주식수'] = sample5['상장주식수'] * 1000
```

```python
sample5['순자산'] = sample5['자산총계'] - sample5['부채총계']
sample5['현재가치'] = sample5['순자산'] / sample5['상장주식수']
```

```python
# 산출한 현재가치보다 현재가가 낮은 경우만 추출
sample6 = sample5[sample5['현재가'] < sample5['현재가치']]
sample6.head()
```

```python
lst = list(set(sample6['종목명'].values))
lst
```

## 1 & 2 교집합으로 저평가 지표 선별

```python
seett = set(sample5['종목명'].values)&set(lst)
```

```python
print(len(seett))
for i in seett:
    print(i,end="  ")
```

```python
b = list(seett)
```

## 3. 교집합 종목명에 점수 부여

```python
df = stock[stock['종목명'].isin(b)]
import numpy as np
df1 = df[['종목명','PER', 'PBR', 'ROE']]
df1 = df1[(df1['PER']>0) & (df1['PBR']>0) & (df1['ROE']>0)]
df1 = df1.sort_values('PER', ignore_index=True)
df1['PER_Score'] = list(range(len(df1), 0, -1))
df1
df1 = df1.sort_values('PBR', ignore_index=True)
df1['PBR_Score'] = len(df1) - np.array(df1.index)
df1
df1 = df1.sort_values('ROE', ascending=False,ignore_index=True)
df1['ROE_Score'] = len(df1) - np.array(df1.index)
df1
df1['total_score'] = df1['PER_Score'] + df1['PBR_Score'] + df1['ROE_Score']
df1 = df1.sort_values('total_score', ascending=False, ignore_index=True)
df1
top_df = []
for i in b:
    top_df.append(df1[df1['종목명']==i])
top_df = pd.concat(top_df, ignore_index=True)
top_df
top_df = top_df.sort_values('total_score', ascending=False)
top_df.head(10)
top_df.head(10)['종목명'].values
```
