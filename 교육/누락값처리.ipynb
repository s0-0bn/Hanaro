```python
import pandas as pd
```

```python
# 기존의 데이터랑 동일한데
# 일부로 누락값을 만들었음

df = pd.read_csv("실습 파일/bike_rentals.csv")
df.info()
```

```python
df.describe()
```

```python
print(df.isna().sum().sum())
print('--------------------------')
print(df.isna().sum())
```

```python
# 누락값만 있는 행만 추출
df[df.isna().any(axis=1)]
```

```python
df['windspeed'] = df['windspeed'].fillna(df['windspeed'].median())
```

```python
df["hum"]=df['hum'].fillna(df.groupby('season')['hum'].transform("median"))
df
```

```python
df['temp'] = df['temp'].interpolate()
df['atemp'] = df['atemp'].interpolate()
```

```python
df
```

```python
df['dteday'] = pd.to_datetime(df['dteday'])
df['mnth'] = df['dteday'].dt.month
df
```

```python
df.loc[730, "yr"] = 1
df
```

```python
df2 = df.drop(['dteday','registered','casual'], axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np

X = df2.drop('cnt', axis=1)
y = df2['cnt']

train_x, test_x, train_y, test_y = train_test_split(X,y)

# 원래 표준화해야하지만 고냥 일단 진행하는 것
lr = LinearRegression()
lr.fit(train_x, train_y)

pred = lr.predict(test_x)

#MSE 값 구하기
np.mean((pred-test_y)**2)
```

```python
# 교차 검증 진행

from sklearn.model_selection import cross_val_score

# 높은 오차를 만들어가는 방향으로 가기 때문에 mean_squared_error 가 아닌
# neg_mean_squared_error
# 앞에 negative 로 가야함! 회귀모델 만들때
model = cross_val_score(lr,X,y, cv=10, scoring="neg_mean_squared_error")

# 음수로 찾았으니, 앞에 "-" 붙여주기
print(-model)

# 5개의 평균으로 보기
# RMSE
np.mean((-model))**(1/2)
```

```python
# 랜덤포레스트 < XGBoost 정형데이터에서는 80~90% 이 모델이 좋다라고 볼 수 있음 (나중에 배워보장)
#(비정형데이터 - 딥러닝)
# 성능좋고 사용하기 좋은데

from xgboost import XGBRegressor

xg = XGBRegressor()

scores = cross_val_score(xg, X,y,scoring='neg_mean_squared_error', cv=10)
```

```python
# 오차가 너무 크게 나오긴 했지만,,, 
# 이게 더 좋은거긴한데...
rmse = (-scores)**(1/2)
rmse.mean()
```
